#### Classes and Objects

Object Oriented Programming (OOP) is a programming paradigm that uses "objects" to design applications and computer programs. OOP alows for modeling real-world scenarios using and objects. This lesson covers the basics of creating classes and objects, including instance variables and methods.

```python
### A class is a blue print for creating objects. Attributes, methods

class Car:
    pass

audi=Car()
bmw=Car()

print(type(audi))
```

```python
## Instance variables 
class Dog:
    ## This is a Constructor(__init__())
    def __init__(self,name,age): 
        ## self is a keyword  used for accesing thr instance variables inside the class itself whenever we create an object.


        ## HERE WE ARE CREATING THE INSTANCE VARIABELS INSIDE THE CONSTRUCTOR ONLY.
        self.name=name
        self.age=age
        ## Self.name is nothing but the instance varibale inside the dog class

## CREATE OBJECTS
dog1 = Dog("Buddy",3) 
## As soon as you call this object ,it will call the constructor and then we ahve to pass the paramenters in it

print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
## we can create multiple objects with the same constructor

dog2= Dog("Lucy",4)
print(dog2.name)
print(dog2.age)
```

```python
## INSTANCE METHODS(FUNCTIONS OR GIVING FUNCTIONALITY TO THE METHODS)

## define a class with instance methods
class Dog:
    def __init__(self,name,age): ## It is a property
        self.name=name
        self.age=age

    def bark(self): ## self keyword is used so that you can use instance variabel inside this method
        print(f"{self.name} says woof")

dog1=Dog("Buddy",4)
dog1.bark()   ## Here i can use this object to call this method

dog2=Dog("Lucy",5)
dog2.bark()

```

```python
## Modeling a bank account

## DEFINE THE BANK ACCOUNT
class BankAccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+= amount
        print(f"{amount} is deposited .New balance is {self.balance}")  

    def withdraw(self,amount):
        if amount >self.balance:
            print("Insufficient funds!")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn. New balance is {self.balance}")    

    def get_balance(self):
        return self.balance
    
## Crating the account

account=BankAccount("Rishi",9000)
print(account.balance)


```

```python
## Call instance methods
account.deposit(100)

account.withdraw(300)

print("The current balance is :",account.get_balance())
```
