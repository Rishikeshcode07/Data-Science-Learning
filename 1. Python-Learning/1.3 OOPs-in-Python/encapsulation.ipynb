#### Encapsulation

Encapsulation is the concept of wrapping data(variables) and maethods (functions) together as a single unit. It restricts direct access to some of the object's components, which is a means of preventing accidental interference and misuse of the data.


```python
## Encapsulation with getter and setter methods

### Public ,protected and private variables or access modifiers

class Person:
    def __init__(self,name,age):
        self.name=name   ## public variables(can be called outside of the specific class)
        self.age=age     ## public variables


## You can see here that you can access the public variables outside the class also.
def get_name(person):
    return person.name

person=Person("Rishi",32)
get_name(person)

```

```python
## Now we will see how we can make the variables private.
## TO make the vairable private we have to use '__'with the variable.
## By making the variable private we cannot access the variable outside the class.

class Person:
    def __init__(self,name,age,gender):
        self.__name=name   ## private variables(cannot be called outside of the specific class)
        self.__age=age     ## private variables
        self.gender=gender ## Public varibale 

def get_name(person):
    return person.__name

person=Person("Rishi",25,"Male")
get_name(person)

## So if you go and print the value then you can't be able to access it as it is private variable.
## But if you change the public to private then you can access it.
```

```python
## Protected Variable
## It says that you cannot access the variable outside the class
## But you can access this via derrived class(inheritance use).

class Person:
    def __init__(self,name,age,gender):
        self._name=name   ## protected variables(cannot be called outside of the specific class)
        self._age=age     ## protected variables
        self.gender=gender ## Public varibale 

class Employee(Person):  ## Derived class
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)


employee=Employee("Rishi",34,"Male")
print(employee._name)
```

```python
## ENCAPSULATION WITH GETTER AND SETTER
class Person:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age

    ## Getter method for name
    def get_name(self):
        return self.__name

    ## Setter method (this will help to change the values in the variable)
    def set_name(self,name):
        self.__name=name    

    ## Getter method for age
    def get_age(self):
        return self.__age
    
    ## Setter method for age
    def set_age(self,age):
        if age > 0:
            self.__age=age
        else:
            print("Age cannot be negative.")    
    

person=Person("Rishi",25)

print(person.get_name())
print(person.get_age())

person.set_age(24)
print(person.get_age())

person.set_age(-5)

```
