### Inheritance In Python

Inheritance is afundamental concept in OOPs that allows a class to inherit attributes and methods afrom another class. This lesson covers single inheritance and multiple inheritance,demonstrating how to create and use them in Python.

```python
                       ## IHERITANCE
                    ## Single Inheritance
#1. Defineing the parent class

class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def drive(self):
        print(f" The person will drive the {self.enginetype} car")

car1=Car(4,5,"Petrol")
car1.drive()        
```

```python
''' 
So now what i want is that i don't want to recreate the 
enitre class again for the properties and methods of a car, so what i 
can do is that i can inherite the properties of the car and
add it to the tesla class. So to do so we have something called 
heritetance . This will inherte the properties of the parent 
class into the subclass.

'''
class Tesla (Car): ## This is how you interite the child class from the parent class
    def __init__(self,windows,doors,enginetype,is_selfdriving):

        ## Super() keyword is used to call the parent class and from the parent class
        ## go and call the __init__ method() i.e nothing but the constructor with the attributes
        super().__init__(windows,doors,enginetype)

        self.is_selfdriving = is_selfdriving
        ##so this is the new feature that i am adding here 
        ## thats way i have to create a new instance of it(i.e properties)


    def selfdriving(self):  ## instance method
        print(f" Tesla supports self driving {self.is_selfdriving}")

tesla1=Tesla(4,5,"electric",True)
tesla1.selfdriving()
tesla1.drive()
```

```python
                      ## Mutiple Ingeritance

## When a class inherits from more than one base class.

class Animal:
    def __init__(self,name):
        self.name=name

    def speak(self):
        print("Subclass must implement this method")

## Base class 2
class Pet:
    def __init__(self,owner):
        self.owner=owner


## Derived class
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)    

    def speak(self):
        return f"{self.name} says woof"

## Create an object
dog=Dog("buddy","rishi")
print(dog.speak())   
print(f"Ower:{dog.owner}")                       
```
