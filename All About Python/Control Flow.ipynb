### CONTROL FLOW



```python
## IF Statement

age=20
if age>=18:
    print("You are allowed to vote in the elections")
```

```python
## Else Statement

age=16

if age>=18:
    print("You are eligible for voting")
else:
    print("You are a minor")    
```

```python
## Elif Statement
## The elif statemnet allows you to check multiple conditions. It stands for "else if"

age= 20
if age<=13:
    print("You are a child")
elif age<18:
    print("You are a teenager")
else:
    print("You are an adult")        

```

```python
## Nested conditional statement

num= int(input("Enter the number"))
if num > 0:
    print("The number is positive")
    if num%2==0:
        print("The number is even")
    else:
        print("The number is odd")
else:
    print("The number is zero or negative")
```

```python
## FOR LOOPS

for i in range(5):
    print(i)
```

```python
for i in range(0,6):
    print(i)
```

```python
for i in range(0,10,2):
    print(i)

## here the range(0,10,2)
## the 3rd part is for increamenting the itteration by 2 steps.
```

```python
for i in range(10,1,-3):
    print(i)
```

```python
## STRINGS

str = "Rishi ojha"
for i in str:
    print(i)
```

```python
## WHILE LOOP

count = 0
while count < 5:
    print(count)
    count = count+1
    
```

```python
count =0
while count % 2 ==0:
    print(count)
    count= count + 1
```

```python
## LOOP CONTROL STATEMENTS
for i in range(10):
    if i==5:
        break
    print(i)

## break is used to come out from the loop as soon the condtions is true    
```

```python
## Continue Statement
## The continue statement skips the current iteration and continues with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## Pass statement
## The pass statement is a null operation and it does nothing.

for i in range(5):
    if i==3:
        print("The number is",i)
        pass
    print(i)
```

```python
## NESTED LOOPS

for i in range(3):
    for j in range(2):
        print(f" i:{i} and j:{j}")
## This is the another way to display to variables in f string format


```
