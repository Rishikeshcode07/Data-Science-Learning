```python
## Type conversion

age=25
print(type(age))

age_str=str(age) ## this is converstion from int to string 
print(age_str)
print(type(age_str))

## other examples are 
# Float to Integer
x = 5.7
print(int(x))  # Output: 5

# String to Integer
y = "123"
print(int(y))  # Output: 123

# Integer to Float
x = 10
print(float(x))  # Output: 10.0

# String to Float
y = "3.14"
print(float(y))  # Output: 3.14

# Integer to String
x = 42
print(str(x))  # Output: "42"

# Float to String
y = 3.14
print(str(y))  # Output: "3.14"

# Integer to Boolean
x = 0
print(bool(x))  # Output: False

# String to Boolean
y = "hello"
print(bool(y))  # Output: True

# Integer to Complex
x = 5
print(complex(x))  # Output: (5+0j)

# String to Complex
y = "3+4j"
print(complex(y))  # Output: (3+4j)

# String to List
x = "abc"
print(list(x))  # Output: ['a', 'b', 'c']

# Tuple to List
y = (1, 2, 3)
print(list(y))  # Output: [1, 2, 3]

# List to Tuple
x = [1, 2, 3]
print(tuple(x))  # Output: (1, 2, 3)

# String to Tuple
y = "xyz"
print(tuple(y))  # Output: ('x', 'y', 'z')

# List to Set
x = [1, 2, 2, 3]
print(set(x))  # Output: {1, 2, 3}

# String to Set
y = "hello"
print(set(y))  # Output: {'h', 'e', 'l', 'o'}

# List of tuples to Dictionary
x = [("a", 1), ("b", 2)]
print(dict(x))  # Output: {'a': 1, 'b': 2}

# List of lists to Dictionary
y = [["x", 10], ["y", 20]]
print(dict(y))  # Output: {'x': 10, 'y': 20}

x = "A"
print(ord(x))  # Output: 65


x = 65
print(chr(x))  # Output: A


x = 10
print(bin(x))  # Output: '0b1010'


x = 255
print(hex(x))  # Output: '0xff'


x = 8
print(oct(x))  # Output: '0o10'


```

```python
## Dynamic typing
## Python allows the type of a variable to change as the program executes

var=10
print(var,type(var))

var="Rishikesh"
print(var,type(var))

var=3.14
print(var,type(var))
```
